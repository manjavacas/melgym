```python
import matplotlib.pyplot as plt
import pandas as pd
from joypy import joyplot

# cvs = ['CV001', 'CV006', 'CV011', 'CV026', 'CV083', 'CV086']
cvs = ['CV015', 'CV023', 'CV024', 'CV054', 'CV080', 'CV081', 'CV082', 'CV089', 'CV096']

def read_data(filename, columns=None):
    if columns is None:
        columns = ['Time'] + cvs
    df = pd.read_csv(filename, delim_whitespace=True, header=None, names=columns)
    return df

filename = 'melgym/out/td3-lr001-b2/PRESSURES.DAT'

columns_to_show = cvs

df = read_data(filename)

df_melted = df.melt(id_vars=['Time'], var_name='CV', value_name='Pressure')

print(df_melted)
```

```python

plt.figure(figsize=(12, 8))
joyplot(
    data=df_melted,
    by="CV",
    column="Time",
    xlabels=True,
    ylabels=True,
    overlap=0.5,  # Reduce el solapamiento para mejorar la visibilidad
    linewidth=1.5,  # Ajusta el grosor de las líneas
    alpha=0.7  # Ajusta la transparencia de las curvas
)

# Ajustes y mostrar la gráfica
plt.title('Evolución de los valores de presión de cada CV a lo largo del tiempo')
plt.xlabel('Tiempo')
plt.ylabel('CVs')
plt.show()
```
